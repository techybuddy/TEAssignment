```python
def gcd(a,b):
    if b==0 :
        return a
    return gcd(b, a%b)


def pour(toJugCap, fromJugCap,d):
    print(toJugCap, "liter capacity jug , ", fromJugCap, "litre capacity jug\n")
    fromJug = fromJugCap
    toJug = 0
    step = 1
    
    print('(', toJug ,',',fromJug,')\n')
    while((fromJug is not d) and (toJug is not d)):
        temp = min(fromJug, toJugCap-toJug)
        toJug += temp
        fromJug -=temp
        
        step += 1
        print('(', toJug ,',',fromJug,')\n')
        
        if((fromJug == d) or (toJug==d)):
            break
            
        if (fromJug == 0):
            fromJug = fromJugCap
            print('(', toJug ,',',fromJug,')\n')
            step += 1
            
        if (toJug == toJugCap):
            toJug = 0
            print('(', toJug ,',',fromJug,')\n')
            step += 1
    return step

    
def minSteps(n,m,d):
    if (d % gcd(n,m) != 0):
        print('-1')
        
    print("\nStep in pouring water from ",m ," litre capacity jug to ",n," litre capacity jug\n")
    s1 = pour(n,m,d)
    print("Total steps : ",s1)
        
    print("\nStep in pouring water from ",n ," litre capacity jug to ",m," litre capacity jug\n")
    s2 = pour(m,n,d)
    print("Total Steps : ",s2)
        
    if(s1 < s2):
        print('\nfrom about we can say that Pouring water from', m, 'Liter capacity Jug to', n, 'liter capacity jug takes minimum steps')
        return
    print('\nfrom about we can say that Pouring water from', n, 'Liter capacity Jug to', m, 'liter capacity jug takes minimum steps')
    return
        
n= int(input('enter capacity of A: '))
m= int(input('enter capacity of B: '))
d= int(input('Goal: '))

minSteps(n, m, d)
```
